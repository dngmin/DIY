#### 컨택트

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',None)

x = np.arange(1,101,1)
y = 5 + (1 / (1 + np.exp(-x/50)) * 210 - 100)*0.9

def data(x):
    print(f'ability {x} :  {(5 + (1 / (1 + np.exp(-x/50)) * 210 - 100)*0.9)}')

data(1)
data(50)
data(100)

x = x.tolist()
y = y.tolist()

lst = {
    'ability' : x,
    'percentage' : y
}

plt.plot(x,y)
plt.title('ability vs batting eye & contact')
plt.xlabel('ability')
plt.ylabel('batting eye & contact (%)')
plt.grid()
plt.show()

df = pd.DataFrame(lst)
df = df.set_index('ability')
display(df)
```

#### 구속 vs 컨택확률

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',None)

x = np.arange(0,200,1)
y = (100**(x*0.005) - 25) / 100

def data(x):
    print(f'ability {x} :  {(100**(x*0.005) - 25) / 100}')

data(100)
data(130)
data(140)
data(150)
data(160)
data(170)

x = x.tolist()
y = y.tolist()

lst = {
    'ballspeed' : x,
    'percentage ' : y
}

plt.plot(x,y)
plt.title('ballspeed vs contact')
plt.xlabel('ballspeed (km/g)')
plt.ylabel('contact (%)')
plt.grid()
plt.show()

df = pd.DataFrame(lst)
df = df.set_index('ballspeed')
display(df)
```

#### 블로킹

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',None)

x = np.arange(1,101,1)
y = (1 / (1 + np.exp(-x/70)) * 280 - 127)

def data(x):
    print(f'ability {x} :  {(1 / (1 + np.exp(-x/70)) * 280 - 127)}')

data(1)
data(50)
data(100)

x = x.tolist()
y = y.tolist()

lst = {
    'ability' : x,
    'blocking' : y
}

plt.plot(x,y)
plt.title('ability vs blocking')
plt.xlabel('ability')
plt.ylabel('blocking (%)')
plt.grid()
plt.show()

df = pd.DataFrame(lst)
df = df.set_index('ability')
display(df)
```
